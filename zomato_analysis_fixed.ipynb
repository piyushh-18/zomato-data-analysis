```python
import pandas as pd

df = pd.read_csv("zomato.csv")

df.head()
```

```python
df.info()
```

```python
# checking missing values 

df.isnull().sum()
```

```python
# drop unnecessary columns from dataset 

df = df.drop(['url', 'address', 'phone', 'reviews_list','menu_item'], axis=1)

df.head()
```

```python
# clean rate columns 

df['rate'] = df['rate'].str.replace('/5',' ')
df['rate'] = df['rate'].replace(['NEW','-'],None)
df['rate'] = df['rate'].astype(float)

df['rate'].head()
```

```python
# clean approx_cost

df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',','')
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(float)
```

```python
df['approx_cost(for two people)'].head()
```

```python
# handle dish_liked 
# filling null values with not mentioned

df['dish_liked'] = df['dish_liked'].fillna("Not Mentioned")
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['rate'])
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.shape
```

# Basic Overview Analysis 


```python
# 1. Average Rating 

df['rate'].mean()
```

```python
# 2) Average cost for two 

df['approx_cost(for two people)'].mean()
```

```python
# top 10  locations by restaurants count
 
df['location'].value_counts().head(10)
```

```python
# top rated locations

df.groupby('location')['rate'].mean().round(2).sort_values(ascending=False).head(10)
```

# Feature Engineering

```python
# creste cost category 

def cost_category(x):
    if x <=300:
        return"Budget"
    elif x <= 700:
        return "Mid Range"
    else:
        return "premium"
    
df['cost_category'] = df['approx_cost(for two people)'].apply(cost_category)
```

```python
df['cost_category'].value_counts()
```

```python
# online order impact 

df.groupby('online_order')['rate'].mean().round(2)
```
